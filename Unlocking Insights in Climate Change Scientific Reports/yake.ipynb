```python
import yake
```

```python
import pandas as pd
```

```python
# Load the text file
with open('Test_Chapter_16.txt', 'r') as f:
    text = f.read()

# Specify the language and max length of phrases
language = "en"
max_ngram_size = 3

# Create a YAKE object
kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, top=50)

# Extract keywords
keywords = kw_extractor.extract_keywords(text)

# Create a Pandas DataFrame of the keywords and their scores
df = pd.DataFrame(keywords, columns=['Keyword', 'Score'])

# Save the DataFrame to an Excel file
df.to_excel('keywords_yake.xlsx', index=False)
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
df = pd.read_excel('KeyPhrases_Chapter16.xlsx')

# Convert the DataFrame to a dictionary
keywords_dict = dict(zip(df['Keyword'], df['Score']))

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white')

# Generate the word cloud from the dictionary
wordcloud.generate_from_frequencies(keywords_dict)

# Display the word cloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
